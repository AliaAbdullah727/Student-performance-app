```python
mapping1={'Male': 0 , 'Female': 1,'Other':2}
mapping2={'No': 0 , 'Yes' :1}
mapping3= {'Fair' :0 , 'Good': 1 , 'Poor':2}
mapping4={'High School':0 ,'Bachelor':1, 'Master':2}
mapping5={'Poor':0 , 'Average':1 , 'Good':2}
mapping6={'No': 0 , 'Yes' :1}
df['gender']=df['gender'].map(mapping1)
df['part_time_job']=df['part_time_job'].map(mapping2)
df['diet_quality']=df['diet_quality'].map(mapping3)
df['parental_education_level']=df['parental_education_level'].map(mapping4)
df['internet_quality']=df['internet_quality'].map(mapping5)
df['extracurricular_participation']=df['extracurricular_participation'].map(mapping6)
df.head()
```

```python
# split the data
df=df.drop(columns=['student_id'])
x=df.iloc[:, :-1]
y=df['exam_score']
x.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.metrics import r2_score , mean_squared_error, accuracy_score
from sklearn.linear_model import LinearRegression , Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train , x_test , y_train , y_test= train_test_split( x, y , random_state=42 , test_size=.2)
x_train.shape , x_test.shape , y_train.shape, y_test.shape
```

```python
# Model 1 (Linar Regression)
lr= LinearRegression()
lr
```

```python
# Model 2 (Random Forest)
rf=RandomForestRegressor()
rf
```

```python
# Model 3 (Decision Tree)
dt=DecisionTreeRegressor(max_depth=5, min_samples_split=20)
dt
```

```python
def evaluate_model(model):
    model.fit(x_train, y_train)
    
    y_pred=model.predict(x_test)
    r2= r2_score( y_test , y_pred)
    MSE= mean_squared_error( y_test , y_pred)
    return {'Mean Squared Error':MSE, 'r2_score': r2}
    
```

```python
linear_regression= evaluate_model(lr)
linear_regression
```

```python
random_forest=evaluate_model(rf)
random_forest
```

```python
desicion_tree=evaluate_model(dt)
desicion_tree
```
